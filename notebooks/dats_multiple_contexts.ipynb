```
"@context": [ "https://w3id.org/dats/context/sdo/dataset_sdo_context.jsonld", "https://w3id.org/dats/context/obo/dataset_obo_context.jsonld"],
```

What is important to know is how JSON-LD interprets these context files, knowing that "duplicate context terms are overridden using a most-recently-defined-wins mechanism." (see [the JSON-LD specification](https://json-ld.org/spec/latest/json-ld/#advanced-context-usage))

For that reason, as we are including the schema.org context file first and the OBO Foundry context files second, if there are OBO Foundry terms mapped to the same elements than the schema.org terms, these will overwrite the schema.org terms. Thus, we are givin priority to OBO Foundry terms. 

Let's now load the topmed file as a graph:

````python
g = Graph().parse(data=topmed, format='json-ld')
````

Now we can visualize the triples in this graph, by checking the graph in Notation3 format:

````python
print g.serialize(format='n3', indent=4)
````

As you can see in the above triples, the DATS dataset is now represented relying on OBO Foundry terms (such as  [obo:IAO_0000100](http://purl.obolibrary.org/obo/IAO_0000100) that represents a 'data set') and also schema.org terms (such as [sdo:AnatomicalStructure](https://schema.org/AnatomicalStructure).
